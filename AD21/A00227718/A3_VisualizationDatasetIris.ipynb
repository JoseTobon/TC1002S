# Importing libraries

```python
# Import the packages that we will be using
import pandas as pd                 # For data handling
import seaborn as sns               # For advanced plotting
import matplotlib.pyplot as plt     # For showing plots
```

# Importing data

```python
# url string that hosts our .csv file
url = "Notebooks/datasets/iris/iris.csv"

# Read the .csv file and store it as a pandas Data Frame
df = pd.read_csv(url, names=["sepal length","sepal width","petal length","petal width","class"])
```

```python
hist = sns.histplot(df["sepal length"], kde = False)
hist = sns.histplot(df["sepal width"], kde = False)
hist = sns.histplot(df["petal length"], kde = False)
hist = sns.histplot(df["petal width"], kde = False)
hist.set_ylabel("cm")
hist.set_xlabel("flowers width and height for sepal and petal")
hist.set_title("Quantitive variables of flowers")
plt.show()
```

```python
hist = sns.histplot(df["sepal length"], kde = False)
hist.set_ylabel("cm")
hist.set_title("Quantitive variables of flowers")
plt.show()
hist = sns.histplot(df["sepal width"], kde = False)
hist.set_ylabel("cm")
hist.set_title("Quantitive variables of flowers")
plt.show()
hist = sns.histplot(df["petal length"], kde = False)
hist.set_ylabel("cm")
hist.set_title("Quantitive variables of flowers")
plt.show()
hist = sns.histplot(df["petal width"], kde = False)
hist.set_ylabel("cm")
hist.set_title("Quantitive variables of flowers")
plt.show()
```

```python
boxplots = df.loc[:, ["sepal length", "sepal width", "petal length", "petal width"]]

box = sns.boxplot(data=boxplots, orient="h")
box.set_xlabel("cm")
box.set_title("Quantitive variables")
plt.show()
```

```python
sns.boxplot(x = df["petal width"], y = df["class"])

plt.show()
```
