```python
# Import the packages that we will be using
import pandas as pd                 # For data handling
import seaborn as sns               # For advanced plotting
import matplotlib.pyplot as plt     # For showing plots

# Note: specific functions of the "sklearn" package will be imported when needed to show concepts easily
```

```python
# url string that hosts our .csv file
# url string that hosts our .csv file
url = "datasets/iris/iris.csv"

# Read the .csv file and store it as a pandas Data Frame
df = pd.read_csv(url, names= ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width', 'Class'])
```

```python
df.shape
```

```python
df = df.dropna()
df.head(15)
```

```python
# Encoding the categorical column
df = df.replace({"Class":  {"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2} })

df
```

## Clustering of petal

```python
# Drop out non necesary columns
df.drop(["Sepal_length","Sepal_width"],axis='columns',inplace=True)

#Visualize the dataset
df
```

```python
plt.scatter(df.Petal_length, df.Petal_width, s=50)
plt.title('Petal length vs. Petal width')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```

```python
df1 = df[df.Class==0]
df2 = df[df.Class==1]
df3 = df[df.Class==2]

plt.scatter(df1.Petal_length, df1.Petal_width, s=50)
plt.scatter(df2.Petal_length, df2.Petal_width, s=50)
plt.scatter(df3.Petal_length, df3.Petal_width, s=50)

plt.title("Scatter plot of my Dataset")
plt.xlabel("Petal length")
plt.ylabel("Petal width")

plt.show
```

```python
# Import sklearn KMeans
from sklearn.cluster import KMeans

# Define number of clusters
km = KMeans(n_clusters=3)

# Do K-means clustering (assing each point in the dataset to a cluster)
clusters = km.fit_predict(df[["Petal_length","Petal_width"]])

# Print estimated cluster of each point in the dataser
clusters
```

```python
# Add a new column to the dataset with the cluster information
df["Cluster_1"] = clusters
df
```

```python
df.Cluster_1.unique()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.Cluster_1==0]
df2 = df[df.Cluster_1==1]
df3 = df[df.Cluster_1==2]

plt.scatter(df1.Petal_length, df1.Petal_width, s=50)
plt.scatter(df2.Petal_length, df2.Petal_width, s=50)
plt.scatter(df3.Petal_length, df3.Petal_width, s=50)

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='magenta', marker='*', label='Centroides', s=256)

plt.title("Scatter plot of my Dataset")
plt.xlabel("Petal length")
plt.ylabel("Petal_width")
plt.legend()

plt.show
```

```python
# Selecting K: elbow plot
sse =[]
k_rng=range(1,10)
for k in k_rng:
    km =KMeans(n_clusters=k)
    km.fit_predict(df[["Petal_length","Petal_width"]])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)

plt.title("elbow plot")
plt.xlabel("K")
plt.ylabel("SSE")

plt.show()
```

- Normalizing the data: preprocessing using min max scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(df[['Petal_length']])
df['Petal_length'] = scaler.transform(df[['Petal_length']])

scaler.fit(df[['Petal_width']])
df['Petal_width'] = scaler.transform(df[['Petal_width']])

df
```

```python
df.describe()
```

```python
plt.scatter(df.Petal_length,df.Petal_width, s=50)
plt.title('Petal_length and Petal_width')
plt.xlabel('Petal_length')
plt.ylabel('Petal_width')
plt.show()
```

```python
# Define number of clusters
km = KMeans(n_clusters=3)

# Do K-means clustering (assing each point in the dataset to a cluster)
clusters = km.fit_predict(df[["Petal_length","Petal_width"]])

# Print estimated cluster of each point in the dataser
clusters
```

```python
# Add a new column to the dataset with the cluster information
df["Cluster_2"] = clusters
df
```

```python
df1 = df[df.Cluster_2==0]
df2 = df[df.Cluster_2==1]
df3 = df[df.Cluster_2==2]

plt.scatter(df1.Petal_length, df1.Petal_width, s=50)
plt.scatter(df2.Petal_length, df2.Petal_width, s=50)
plt.scatter(df3.Petal_length, df3.Petal_width, s=50)

plt.title('Petal_length and Petal_width')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```

```python
df = pd.read_csv(url, names= ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width', 'Class'])
df = df.dropna()

# Encoding the categorical column
df = df.replace({"Class":  {"Iris-setosa":0, "Iris-versicolor":1, "Iris-virginica":2} })

# Define number of clusters
km = KMeans(n_clusters=3)

# Do K-means clustering (assing each point in the dataset to a cluster)
clusters = km.fit_predict(df[['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width']])

df["Cluster_1"] = clusters

print(clusters)
print(df)
print(km.cluster_centers_)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(df[['Sepal_length']])
df['Sepal_length'] = scaler.transform(df[['Sepal_length']])

scaler.fit(df[['Sepal_width']])
df['Sepal_width'] = scaler.transform(df[['Sepal_width']])

scaler.fit(df[['Petal_length']])
df['Petal_length'] = scaler.transform(df[['Petal_length']])

scaler.fit(df[['Petal_width']])
df['Petal_width'] = scaler.transform(df[['Petal_width']])

df
```
