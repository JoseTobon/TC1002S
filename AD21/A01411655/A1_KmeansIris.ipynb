# Activity: work with the iris dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans

url = "datasets/iris/iris.csv"
df = pd.read_csv(url)

df
```

1. Do clustering with the iris flower dataset to form clusters using as features petal width and length. Drop out the other two features (sepal width and length) for simplicity.

```python
df_petal = df.loc[:,["PetalL","PetalW","Class"]]
classRepl = {"Class": {"Iris-setosa":1, "Iris-versicolor":2, "Iris-virginica":3}}
df_petal = df_petal.replace(classRepl)
km = KMeans(n_clusters=3)
yp = km.fit_predict(df_petal[["PetalL","PetalW"]])
yp
```

```python
df_petal["Cluster1"]=yp
df_petal
```

```python
df_petal.Cluster1.unique()
```

```python
df1 = df_petal[df_petal.Cluster1==1]
df2 = df_petal[df_petal.Cluster1==0]
df3 = df_petal[df_petal.Cluster1==2]

plt.scatter(df1.PetalL, df1.PetalW, Label="Estimated type 1")
plt.scatter(df2.PetalL, df2.PetalW, Label="Estimated type 2")
plt.scatter(df3.PetalL, df3.PetalW, Label="Estimated type 3")

plt.title('Petal length vs Petal width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(df[['SepalL']])
df['SepalL'] = scaler.transform(df[['SepalL']])

scaler.fit(df[['SepalW']])
df['SepalW'] = scaler.transform(df[['SepalW']])

scaler.fit(df[['PetalL']])
df['PetalL'] = scaler.transform(df[['PetalL']])

scaler.fit(df[['PetalW']])
df['PetalW'] = scaler.transform(df[['PetalW']])

df
```

```python
km = KMeans(n_clusters=3)
yp = km.fit_predict(df[["SepalL","SepalW","PetalL","PetalW"]])
yp
```

    - Compute and show the elbow plot

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit_predict(df[["SepalL","SepalW","PetalL","PetalW"]])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.show()
```
