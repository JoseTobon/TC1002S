# Kmeans clustering

Kmeans clustering

```python
#import machine learning library
from sklearn.cluster import KMeans

k = 3 #always define this k
#create object to do kmeans and distribute clusters
Km = KMeans(n_clusters=k)

#do kmeans
clusters = Km.fit_predict(irisData[['SepalLength','SepalWidth','PetalLength', 'PetalWidth']]) #this is formula
#!note distance between points is always the same ecuation = sqrt(dx1^2 + dx2^2 + dx3^2 + ... dxn^2)

clusters
```

```python
#insert one new column to compare clusters vs flower type
irisData['cluster1'] = clusters
irisData
#we realize that: flower 1 == cluster 0, flower 2 == cluster 2 and flower 3 == cluster 1 or likewhise, it still has some errors
```

```python
irisData.cluster1.unique()
```

```python
Km.cluster_centers_
```

```python
Km.inertia_
```

<span style='color:Blue'> **Important remarks**  </span>

- The number of each cluster is randomly assigned
- The order of the number in each cluster is random

# Plot estimated clusters

Plot estimated clusters

```python
#since we cannot plot we will compare flower vs cluster
#first we fix the Flower type acording to the index from the clusters without altering the flower type so that the comparison is more alike
irisData = irisData.replace({"cluster1": {irisData.cluster1.unique()[0]:0,irisData.cluster1.unique()[1]:1,irisData.cluster1.unique()[2]:2} })
print((irisData['Flower'] == irisData['cluster1']).sum())
#we can see that theres a good approximation
```

# Selecting K: elbow plot

Check the acurracy of the model using k-fold cross-validation

```python
#sum of squared distances
sse = []
#k ranges
k_rng = range(1,10)

for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit_predict(irisData[['SepalLength','SepalWidth','PetalLength', 'PetalWidth']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('SSE')

plt.show
```

<span style='color:Blue'> **Important remarks**  </span>

According to the Elbot plot, the selected K agree with the real number of clusters



# Normalizing the data: preprocessing using min max scaler

Normalizing the data: preprocessing using min max scaler

```python
from sklearn.preprocessing import MinMaxScaler
#create object
scaler = MinMaxScaler()
#normalizing data for all four variables
scaler.fit(irisData[['SepalLength']])
irisData['SepalLength'] = scaler.transform(irisData[['SepalLength']])

scaler.fit(irisData[['SepalWidth']])
irisData['SepalWidth'] = scaler.transform(irisData[['SepalWidth']])

scaler.fit(irisData[['PetalLength']])
irisData['PetalLength'] = scaler.transform(irisData[['PetalLength']])

scaler.fit(irisData[['PetalWidth']])
irisData['PetalWidth'] = scaler.transform(irisData[['PetalWidth']])

irisData
```
