<h4>Primero obtenemos una vista general de los datos a representar.

```python
# Import the packages that we will be using
import pandas as pd                 # For data handling
import seaborn as sns               # For advanced plotting
import matplotlib.pyplot as plt     # For showing plots

# url string that hosts our .csv file
url = "datasets/iris/iris.csv"

# Read the .csv file and store it as a pandas Data Frame
dt = pd.read_csv(url)

dt.shape
dt.head()
```

<h3>1. Plot the histograms for each of the four quantitative variables

```python
h1 = sns.histplot(dt["sepal length"], kde = False)
h1 = sns.histplot(dt["sepal width"], kde = False)
h1 = sns.histplot(dt["petal length"], kde = False)
h1 = sns.histplot(dt["petal width"], kde = False)
h1.set_title("Histogram of quantitative variables")
h1.set_ylabel("Measurements in centimeters")
h1.set_xlabel("Sepal and petal width and length measurements")
plt.show()
```

<h3>2. Plot the histograms for each of the quantitative variables

```python
h1 = sns.histplot(dt["sepal length"], kde = False)
h1.set_title("Histogram of quantitative variables")
h1.set_ylabel("Measurements in centimeters")
plt.show()
h1 = sns.histplot(dt["sepal width"], kde = False)
h1.set_title("Histogram of quantitative variables")
h1.set_ylabel("Measurements in centimeters")
plt.show()
h1 = sns.histplot(dt["petal length"], kde = False)
h1.set_title("Histogram of quantitative variables")
h1.set_ylabel("Measurements in centimeters")
plt.show()
h1 = sns.histplot(dt["petal width"], kde = False)
h1.set_title("Histogram of quantitative variables")
h1.set_ylabel("Measurements in centimeters")
plt.show()
```

<h3>3. Plot the boxplots for each of the quantitative variables

```python
bplots = dt.loc[:, ["sepal length", "sepal width","petal length","petal width"]]
b = sns.boxplot(data = bplots, orient = "h")
b.set_xlabel("Measurements in centimeters")
b.set_title("Boxplot of quantitative variables")
plt.show()
```

<h3>4. Plot the boxplots of the petal width grouped by type of flower

```python
sns.boxplot(x = dt["petal width"], y = dt["type"])
plt.show()
```

<h3>5. Plot the boxplots of the setal length grouped by type of flower

```python
sns.boxplot(x = dt["sepal length"], y = dt["type"])
plt.show()
```
