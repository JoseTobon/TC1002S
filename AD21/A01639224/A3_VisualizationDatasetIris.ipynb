# Frequency tables

The `value_counts()` method can be used to determine the number of times that each distinct value of a variable occurs in a data set.  In statistical terms, this is the "frequency distribution" of the variable. The `value_counts()` method produces a table with two columns.  The first column contains all distinct observed values for the variable.  The second column contains the number of times each of these values occurs.  Note that the table returned by `value_counts()` is actually a **Pandas** data frame, so can be further processed using any Pandas methods for working with data frames.

```python
# Number of times that each distinct value of a variable occurs in a data set
df.SepalLength.value_counts()
```

```python
# Proportion of each distinct value of a variable occurs in a data set
x = df.SepalLength.value_counts()
x / x.sum()
```

Note that the `value_counts()` method excludes missing values. We confirm this below by adding up observations to your data frame with some missing values and then computing `value_counts()` and comparing this to the total number of rows in the data set, which is 150. This tells us that there are 150 - 150 = 0 missing values for this variable (other variables may have different numbers of missing values).

```python
# Total number of observations in tips
print(df.shape)

# Total number of null observations in tips
#print(df.isnull(df.Age).sum())

# Total number of counts (excluding missing values) in tips
print(df.SepalLength.value_counts().sum())


```

# Histogram

It is often good to get a feel for the shape of the distribution of the data.

```python
# Plot Sepal Length of each Iris plants
dfhist = sns.distplot(df["SepalLength"], kde = False)
dfhist.set_ylabel("Number of Plants")
dfhist.set_title("Sepal Length in Iris Plants")

plt.show()
```

```python
# Plot distribution of Sepal Length in Iris plants
dfhist = sns.distplot(df["SepalLength"], kde = True)
dfhist.set_ylabel("Number of Plants")
dfhist.set_xlabel("Sepal Length (mm)")
dfhist.set_title("Sepal Length in Iris Plants")

plt.show()
```

```python
# Plot histogram of both the Sepal Length and the Width
sns.distplot(df["SepalLength"], kde = False)
dfhist = sns.distplot(df["SepalWidth"], kde = False)
dfhist.set_ylabel("Number of Plants")
dfhist.set_title("Sepal Length and Width in Iris Plants")

dfhist.set_xlabel("Size (cm)")

plt.show()
```

# Histograms plotted by groups

While looking at a single variable is interesting, it is often useful to see how a variable changes in response to another. Thus, we can create a histograms of one quantitative variable grouped by another categorical variables.

```python
# Create histograms of the "SepalLength" grouped by "Class"
g = sns.FacetGrid(df, row = "Class")
g = g.map(plt.hist, "SepalLength")

plt.show()
```

# Boxplots

Boxplots do not show the shape of the distribution, but they can give us a better idea about the center and spread of the distribution as well as any potential outliers that may exist. Boxplots and Histograms often complement each other and help an analyst get more information about the data

```python
# Create the boxplot of the Sepal Lengths
sns.boxplot(df["SepalLength"]).set_title("Box Plot of the Sepal Lengths")
plt.show()
```

```python
# Create the boxplot of the Sepal Widths
sns.boxplot(df["SepalWidth"]).set_title("Box Plot of the Sepal Widths")
plt.show()
```

```python
# Create the boxplots of the "SepalLength" and of the "SepalWidth" sizes
sns.boxplot(df["SepalLength"])
sns.boxplot(df["SepalWidth"]).set_title("Box Plot of the Sepal Lengths and Widths")
plt.show()

# NOTE: This is not convenient
```

```python
# Create the boxplots of the "SepalLength" and of the "SepalWidth" sizes

SepalLengthAndWidth = df.loc[:, ["SepalLength", "SepalWidth"]]

x2bp = sns.boxplot(data=SepalLengthAndWidth, orient="h")
x2bp.set_xlabel("Size (cm)")
x2bp.set_title("Box Plot of the Sepal Lengths and Widths")

plt.show()
```

# Boxplots plotted by groups

While looking at a single variable is interesting, it is often useful to see how a variable changes in response to another. Thus, we can create a side-by-side boxplots of one quantitative variable grouped by another categorical variables.

```python
# Create side-by-side boxplots of the "SepalLength" grouped by "Class"
sns.boxplot(x = df["SepalLength"], y = df["Class"])
plt.show()
```

# Histograms and boxplots plotted by groups

We cal also create both boxplots and histograms of one quantitative variable grouped by another categorical variables

```python
# Create a boxplot and histogram of the "SepalLengths" grouped by "Class"
sns.boxplot(x = df["SepalLength"], y = df["Class"])

g = sns.FacetGrid(df, row = "Class")
g = g.map(plt.hist, "SepalLength")

plt.show()
```
