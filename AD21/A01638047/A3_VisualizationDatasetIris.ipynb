# Visualizing Data in Python
## Importing libraries

```python
# Import the packages that we will be using
import pandas as pd                 # For data handling
import seaborn as sns               # For advanced plotting
import matplotlib.pyplot as plt     # For showing plots
```

## Importing data

```python
# url string that hosts our .csv file
url = "datasets/iris/iris.csv"

# Read the .csv file and store it as a pandas Data Frame
df = pd.read_csv(url, names=["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Class"])
```

## Exploring the content of the data set

```python
#df
df.shape
df.head()
#df.columns
#df.describe()
```

# Frequency tables

```python
# Number of times that each distinct value of a variable occurs in a data set
df.Class.value_counts()
```

```python
# Proportion of each distinct value of a variable occurs in a data set
x = df.Class.value_counts()
x / x.sum()
```

```python
# Total number of observations in tips
print(df.shape)

# Total number of null observations in tips
#print(df.isnull(df.SepalLength).sum())

# Total number of counts (excluding missing values) in tips
print(df.Class.value_counts().sum())
```

# Histogram

```python
# Plot histogram of the Sepal Length
dfhist = sns.distplot(df["SepalLength"], kde = False)#.set_title("Histogram of Sepal Length")
dfhist.set_ylabel("Frequency")
dfhist.set_title("Histogram of Sepal Length")
plt.show()
```

```python
# Plot distribution of the tips only
dfhist = sns.distplot(df["SepalLength"], kde = True)#.set_title("Distribution of the Sepal Length")
dfhist.set_ylabel("PDF")
dfhist.set_xlabel("Tip")
dfhist.set_title("Distribution of the Sepal Length")

plt.show()
```

```python
# Plot histogram of both the Sepal Length and the SepalWidth
sns.distplot(df["SepalLength"], kde = False)
dfhist = sns.distplot(df["SepalWidth"], kde = False)#.set_title("Histogram of both Age and Wingspan")
dfhist.set_ylabel("Frequency")
dfhist.set_title("Histogram of both Sepal Length and the SepalWidth")

plt.show()
```

# Histograms plotted by groups

```python
# Create histograms of the "SepalLength" grouped by "Class"
g = sns.FacetGrid(df, row = "Class")
g = g.map(plt.hist, "SepalLength")

plt.show()
```

# Boxplots

```python
# Create the boxplot of the Sepal Length
sns.boxplot(df["SepalLength"]).set_title("Box plot of the Sepal Length")
plt.show()
```

```python
# Create the boxplot of the "tips" amounts
sns.boxplot(df["SepalWidth"]).set_title("Box plot of the Sepal Width")
plt.show()
```

```python
# Create the boxplots of the "SepalLength" and of the "SepalWidth" amounts
sns.boxplot(df["SepalLength"])
sns.boxplot(df["SepalWidth"]).set_title("Box plot of the Total Sepal Length and Sepal Width")
plt.show()

# NOTE: this is not convenient
```

```python
# Create the boxplots of the "SepalLength" and of the "SepalWidth" amounts
BillAndTip = df.loc[:, ["SepalLength", "SepalWidth"]]

x2bp = sns.boxplot(data=BillAndTip, orient="h")
x2bp.set_xlabel("cm")
x2bp.set_title("Box plot of the Sepal Length and Sepal Width")

plt.show()
```

# Boxplots plotted by groups

```python
# Create side-by-side boxplots of the "SepalLength" grouped by "Class"
sns.boxplot(x = df["SepalLength"], y = df["Class"])
plt.show()
```

# Histograms and boxplots plotted by groups

```python
# Create a boxplot and histogram of the "tips" grouped by "Class"
sns.boxplot(x = df["SepalLength"], y = df["Class"])

g = sns.FacetGrid(df, row = "Class")
g = g.map(plt.hist, "SepalLength")

plt.show()
```
